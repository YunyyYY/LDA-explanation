```python
# %matplotlib inline
from utils import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from matplotlib import animation, rc
from IPython.display import HTML
```

Citation from [Linear Discriminant Analysis â€“ from Theory to Code](http://www.adeveloperdiary.com/data-science/machine-learning/linear-discriminant-analysis-from-theory-to-code/#more-1141).

```python
# get color attributes
color, carray, cmap = cattr(N=200)
```

```python
# skewed binormal
# get skewed Gaussian distributed data
data_norm = gen_norm(N=200)

pxs, W, m0, m1 = lda(data_norm, color)
fig, ax = plt.subplots(figsize=(5, 5), dpi=150)
ax.axis('equal')
ax.set_xlim((0, 10))
ax.set_ylim((0, 10))

k = W[1]/W[0]
x = np.linspace(0, 10, 10)
y = k * (x-5)+5
ax.plot(x, y, lw=1, alpha=0.3, c='gray')
pxs = np.array(pxs)
scat = ax.scatter(data_norm[:,0], data_norm[:,1], c=color, alpha=0.5, cmap=cmap, s=7)
ax.legend(*scat.legend_elements(num=1),
                    loc="upper right", title="class")
ax.scatter(pxs[:, 0], pxs[:, 1], c=color, cmap=cmap, alpha=0.3, s=7)
ax.plot(2, 5, marker='*', c='royalblue')
ax.scatter(6, 2, marker='*', c='orangered')
plt.savefig('skew_norm_N200.svg')
```
